## Niveles promedio de contaminantes en una zona 

```python
data_ciudad = data_mediciones[(data_mediciones.city.isin(['Monterrey']))]
data_ciudad['fecha_hora']  = pd.to_datetime(data_ciudad.fecha + ' ' +
                                            data_ciudad.hora.astype(str) +':00:00')
print(data_ciudad.shape)
data_ciudad.head()
```

```python
data_estaciones.rename(columns={'id':'estacionesid'}, inplace=True)

data_ciudad.merge(data_estaciones, on='estacionesid').redesid.value_counts()
```

```python
aux = data_ciudad[data_ciudad.parametro=='CO'].sort_values('fecha_hora')

aux.describe()
```

```python
aux = aux[(aux.valororig.between(0, 10))]

fig, ax = plt.subplots(figsize=(14, 7))
p = aux.groupby('estacionesid').plot(x='fecha_hora', y='valororig', ax=ax, legend=False, alpha=0.7)
```

```python
fig, ax = plt.subplots(figsize=(14, 7))

for i, g in data_ciudad.groupby('parametro'):
    if i in ['CO', 'O3', 'SO2', 'NO2']:
        g.groupby('hora').valororig.mean().sort_index().plot(ax=ax, label=i, legend=True)
```

## Ozono en el Valle de México

```python
O3_valleMex = data_mediciones[(data_mediciones.parametro.isin(['O3'])) &
                              (data_mediciones.city.isin(['Valle de México']))]
O3_valleMex['fecha_hora'] = pd.to_datetime(O3_valleMex.fecha + ' ' +
                                           O3_valleMex.hora.astype(str) +':00:00')
O3_valleMex['fecha'] = pd.to_datetime(O3_valleMex.fecha)

O3_valleMex['dia_semana'] = O3_valleMex.fecha.dt.weekday_name

O3_valleMex = O3_valleMex.sort_values('fecha_hora')
print(O3_valleMex.shape)
O3_valleMex.head()
```

```python
data_estaciones[data_estaciones.estacionesid==269]
```

```python
O3_valleMex[O3_valleMex.valororig>0.095]
```

```python
data_estaciones.rename(columns={'id':'estacionesid'}, inplace=True)

O3_valleMex.merge(data_estaciones, on='estacionesid').redesid.value_counts()
```

```python
fig, ax = plt.subplots(figsize=(28, 7))
p = O3_valleMex.groupby('estacionesid').plot(x='fecha_hora', y='valororig', ax=ax, legend=False, alpha=0.7)
```

```python
O3_valleMex.hora.value_counts().sort_index().plot.bar()
```

```python
O3_valleMex.groupby('hora').valororig.mean().plot()
```

```python
O3_valleMex.groupby('dia_semana').valororig.describe()
```

```python
O3_valleMex.groupby('dia_semana').valororig.mean().plot.bar()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

for i, g in O3_valleMex.groupby(['dia_semana']):
    g.groupby('hora').valororig.mean().plot(label=i, legend=True)
plt.show()
```

```python
aux = O3_valleMex[O3_valleMex.estacionesid==259]
aux['fecha_hora'] = pd.to_datetime(aux.fecha_hora)
aux = aux.sort_values('fecha_hora')
aux.head()
```

```python
data = [go.Scatter(x=aux.fecha_hora, y=aux.valororig)]

py.iplot(data, filename='pandas-time-series')
```

```python
aux = O3_valleMex.groupby('fecha').valororig.mean()
aux.index = pd.to_datetime(aux.index)
aux = aux.sort_index()

data = [go.Scatter(x=aux.index, y=aux.values)]

py.iplot(data, filename='pandas-time-series')
```
