## Niveles promedio de ozono por estación

```python
medicion_contaminante = mediciones[mediciones.parametro=='O3']

# De acuerdo al análisis anterior, es buena idea filtrar valores demasiado grandes. Vamos a usar
# como filtro el mínimo valor de ozono para establecer calidad del aire muy mala según SEMARNAT que es 0.6 ppm
medicion_contaminante = medicion_contaminante[medicion_contaminante.valororig.between(0, 0.7)]

medicion_contaminante = medicion_contaminante.groupby('estacionesid').agg({'valororig':['mean', 'count'],
                                                                           'validoorig':'mean'}).reset_index()
medicion_contaminante.columns = ['estacionesid', 'valororig', 'num_registros', 'validoorig']

medicion_contaminante = medicion_contaminante.merge(estaciones[['estacionesid', 'lat', 'long', 'nombre']],
                                                    on='estacionesid')
medicion_contaminante = medicion_contaminante.merge(mediciones[['estacionesid', 'city']].drop_duplicates(),
                                                    on='estacionesid')

print(medicion_contaminante.shape)
medicion_contaminante.head()
```

```python
medicion_contaminante[medicion_contaminante.city=='Monterrey']
```

```python
medicion_contaminante.describe()
```

```python
medicion_contaminante[~medicion_contaminante.valororig.between(0, 0.2)]
```

```python
# Vamos a considerar únicamente estaciones que tengan más de 2400 mediciones por hora.
# Lo que corresponde a 100 días de registros.
mask = (medicion_contaminante.num_registros>2400)

medicion_contaminante[mask].valororig.plot.hist(bins=15)
```

```python
cm = plt.cm.get_cmap('viridis')

mask = (medicion_contaminante.num_registros>2400)
filtered_medicion = medicion_contaminante[mask]

print(filtered_medicion.shape)
filtered_medicion.head()
```

```python
filtered_medicion.describe()
```

```python
plt.figure(figsize=(12, 8))
sc = plt.scatter(filtered_medicion.long, filtered_medicion.lat, c=filtered_medicion.valororig,
                 s=55, cmap=cm)

a = mplleaflet.display()

# Make a figure and axes with dimensions as desired.
fig = plt.figure(figsize=(8, 3))
ax1 = fig.add_axes([0.0, 0.80, 1, 0.15])

# Set the colormap and norm to correspond to the data for which
# the colorbar will be used.
cmap = mpl.cm.viridis
norm = mpl.colors.Normalize(vmin=filtered_medicion.valororig.min(), vmax=filtered_medicion.valororig.max())

# ColorbarBase derives from ScalarMappable and puts a colorbar
# in a specified axes, so it has everything needed for a
# standalone colorbar.  There are many more kwargs, but the
# following gives a basic continuous colorbar with ticks
# and labels.
cb1 = mpl.colorbar.ColorbarBase(ax1, cmap=cmap,
                                norm=norm,
                                orientation='horizontal')
cb1.set_label('Valor promedio de O3')

a
```

```python
# Lugares con peores mediciones promedio
filtered_medicion[filtered_medicion.valororig>filtered_medicion.valororig.max()-filtered_medicion.valororig.std()
                 ].sort_values('valororig')
```

```python
# Lugares con mejores mediciones promedio
filtered_medicion[filtered_medicion.valororig<filtered_medicion.valororig.min()+filtered_medicion.valororig.std()
                 ].sort_values('valororig')
```

```python
for i, g in filtered_medicion.groupby('city'):
    if g.shape[0] >= 4:
        g.valororig.plot.hist(alpha=0.5, label=i, legend=True)
```

```python
_estaciones = filtered_medicion[filtered_medicion.city.isin(['Valle de México', 'Guadalajara',
                                                             'Monterrey', 'Toluca'])].estacionesid

mask = (mediciones.parametro=='O3') & (mediciones.estacionesid.isin(_estaciones))

for i, g in mediciones[mask].groupby('city'):
    g['dia_semana'] = pd.to_datetime(g.fecha).dt.weekday_name
    g.groupby('hora').valororig.mean().plot(figsize=(12, 5), label=i, legend=True)
    
plt.title('Mediciones promedio de Ozono por hora del día por ciudad')
plt.ylabel('ppm')
```

```python
l = []

for i, g in mediciones[mask].groupby('city'):
    g['dia_semana'] = pd.to_datetime(g.fecha).dt.weekday_name
    g['dia_semana_numero'] = pd.to_datetime(g.fecha).dt.weekday
    aux = g.groupby(['dia_semana', 'dia_semana_numero']).valororig.mean().sort_index(level=1).reset_index()
    aux['city'] = i
    l.append(aux[['dia_semana_numero', 'valororig', 'city']])
    
aux = pd.concat(l).pivot('dia_semana_numero', 'city', 'valororig')
aux = aux[['Valle de México', 'Guadalajara', 'Toluca', 'Monterrey']].reset_index()
nombres_dias = {0:'Lunes', 1:'Martes', 2:'Miercoles', 3:'Jueves', 4:'Viernes', 5:'Sabado', 6:'Domingo'}
aux['dia_semana'] = aux.dia_semana_numero.map(nombres_dias)
aux.drop('dia_semana_numero', axis=1, inplace=True)
aux.set_index('dia_semana').plot.bar(figsize=(12, 6))
```

## ¿Qué tanto está más contaminada la ciudad de México?

- Vamos a comparar la ciudad de México con las siguientes 3 ciudades con más estaciones de medición de Ozono para estimar qué tanto más contaminada está la CDMX

```python
# Número de estaciones por ciudad
filtered_medicion.city.value_counts().head(4)
```

```python
mask = filtered_medicion.city.isin(['Valle de México', 'Guadalajara', 'Toluca', 'Monterrey'])
aux = filtered_medicion[mask].groupby('city').agg({'valororig':'mean', 'estacionesid':'nunique'}).reset_index()
aux['escala'] = aux.valororig/aux[aux.city=='Valle de México'].valororig.values
aux = aux[['city', 'estacionesid', 'valororig', 'escala']]
aux.columns = ['ciudad', 'conteo_estaciones_filtradas', 'valor_promedio', 'escala_comparativa']

aux.sort_values('escala_comparativa', ascending=False)
```
